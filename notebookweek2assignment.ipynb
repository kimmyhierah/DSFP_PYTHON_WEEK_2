```python
#14]a function that return grade from score

def grade(score):
    if score >= 70:
        return "A"
    elif score >= 60:
        return "B"
    elif score >= 50:
        return "C"
    elif score >= 45:
        return "D"
    return "F"
```

```python
#21] a function that return factorial of a number

def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
# 27] A function that can check if a number is prime

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
```

```python
print(is_prime(3))
print(is_prime(5))
print(is_prime(2))
print(is_prime(12))
print(is_prime(9))
print(is_prime(14))
```

```python
#29] a function that can return sum of digits of a number

def sum_of_digits(num):
    return sum(int(digit) for digit in str(abs(num)))
```

```python
print(sum_of_digits(1234))
```

```python
#30]a function that can apply default discount to prices

def apply_discount(prices, discount_rate=0.10):
    return [price * (1 - discount_rate) for price in prices]
```

```python
prices = [25000, 77000, 90000]
print(apply_discount(prices))
```
