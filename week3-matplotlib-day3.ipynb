**LINESTYLE**

```python
X = np.array[1,2,3,5,6,8,10]
Y = np.array[4,6,8,19,12,14,18]
```

```python
plt.figure(figsize=(12,10))
plt.title("STUDENTS PERFORMANCE PLOT", fontstyle= "italic" ,fontsize= 20, fontweight= "bold", color= "green") #label
plt.plot(X,Y, marker = "*", 
         markersize = 14, 
         markerfacecolor ='green',
        linestyle= "solid",
        linewidth= 3,
        color= 'pink')
plt.xlabel("READ HOURS", fontsize= 17) #label
plt.ylabel("SCORES", fontsize= 17) #label
plt.grid(linestyle="dashed", color="gray") #grid
plt.figure(figsize=(8,6)) #size..that is the size of your plotting


plt.show()
```

```python
X = np.array([1, 2, 3, 5, 6, 8, 10])
Y = np.array([4, 6, 8, 19, 12, 14, 18])
A = np.array([2, 3, 4, 6, 8, 9, 11])
B = np.array([5, 7, 9, 20, 13, 15, 19])
```

```python
#you need to specify the size first before specifying wat goes inside
plt.figure(figsize=(12,10))
plt.title("STUDENTS PERFORMANCE PLOT", fontstyle= "italic" ,fontsize= 20, fontweight= "bold", color= "green")
plt.plot(X,Y, marker = "*", 
         markersize = 14, 
         markerfacecolor ='green',
        linestyle= "dotted",
        linewidth= 3,
        color= 'pink')

plt.plot(X,A, marker = "+", 
         markersize = 14, 
         markerfacecolor ='orange',
        linestyle= "dotted",
        linewidth= 3,
        color= 'black')

plt.plot(X,B, marker = ".", 
         markersize = 14, 
         markerfacecolor ='purple',
        linestyle= "dotted",
        linewidth= 3,
        color= 'blue')

plt.grid(linestyle="dashed", color="gray")
plt.figure(figsize=(2,1))
plt.show()
plt.savefig("firstplot.png", dpi=300)
```

```python
A= ['Jollof rice', 'Fried rice', 'Waina', 'Swallow']
B= [5,7,10,3]
colors = ['red', 'orange', 'purple', 'green'] #colors for each bar

plt.figure(figsize=(12,10))
plt.title("FOOD AVAILABILITY", fontstyle= "italic" ,fontsize= 20, fontweight= "bold", color= "green")
plt.barh(A,B, color= colors) #horizontal bar plot
plt.grid(linestyle="dashed", color="gray")
plt.xlabel("Food Items", fontsize=16)
plt.ylabel("Quantity Available", fontsize=16)
plt.figure(figsize=(10,8))

plt.show()
```

**PIE CHART**

```python
#Our data
subjects= ['Mathematics', 'Biology', 'Chemistry', 'English', 'Geography']
score= np.array([20,30,40,50,60])
COLORS=['red', 'blue', 'green', 'orange', 'purple']

#plot the pie chart
plt.pie(score, 
        labels=subjects, 
        autopct='%1.1f%%', 
        colors= COLORS,
       explode= [0.1, 0, 0.1, 0, 0.1], #Explode the slice
       shadow= True)

#title to our pie chart
plt.title("STUDENTS SCORES DISTRIBUTION", 
          fontsize=18, 
          fontweight='bold', 
          fontstyle='italic', 
          color='navy')

plt.show()
```

**SCATTER PLOT**

```python
#Our data
X = np.array([1, 2, 3, 5, 6, 8, 10])
Y = np.array([4, 6, 8, 19, 12, 14, 18])

#create the scatter plot
plt.figure(figsize=(8,5))
plt.scatter(X, Y, color='blue')  

# Add title and labels
plt.title("SCATTER PLOT OF X vs Y", fontsize=18, fontweight='bold', fontstyle='italic', color='green')
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```
