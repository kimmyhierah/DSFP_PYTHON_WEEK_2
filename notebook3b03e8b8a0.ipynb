```python
def calculate_bmi(patient_name, weight_kg, height_m):
    bmi = weight_kg / (height_m **2)
    if bmi < 18.5:
        category = "Underweight"
    elif 18.5 <= bmi < 25:
        category = "Normal weight"
    elif 25 <= bmi < 30:
        category = "Overweight"
    else:
        category = "Obese"
    result = {
        "patient_name": patient_name,
        "weight_kg": weight_kg,
        "height_m": height_m,
        "bmi": round(bmi, 2),
        "category": category
    }
    return result
```

```python
result = calculate_bmi("Fatima", 65, 1.65)
print(result)
```

```python
print(calculate_bmi("Myhierah", 63, 2.86))
```

```python
#a function that return the maximum of two numbers

def maximum(a, b):
    if a > b:
        return a
    else:
        return b
```

```python
print(maximum(30, 20))
```

```python
#write a function that takes a list of numbers and return their average 

def average(numbers):
    total = sum(numbers)
    count = len(numbers)
    if count == 0:
        return 0  
    return total / count
```

```python
nums = [10, 20, 30, 40, 50]
print(average(nums))
```

```python
#a function that converts minutes to hours and minutes

def convert_minutes(total_minutes):
    hours = total_minutes // 60  
    minutes = total_minutes % 60 
    return hours, minutes
```

```python
h, m = convert_minutes(90)
print(f"{h} hours and {m} minutes")
```

```python
#a function that takes any number of arguments and returns their sum

def sum_all(*args):
    return sum(args)              
```

```python
print(sum_all(1, 2, 3, 4, 5))  
```

```python
#a function that takes any number of arguments and returns their cumulative product

def product_all(*args):
    result = 1
    for num in args:
        result *= num
    return result
```

```python
print(product_all(2, 3, 4))
```
