```python
def calculate(a, b):
    addition = a+b
    subtraction = a-b
    division = a/b
    multiplication = a*b if b != 0 else "undefined"
    return addition, subtraction, division, multiplication
```

```python
addition, subtraction, division, multiplication = calculate(10, 5)
print(addition,subtraction,division,multiplication)
```
